#### Pie Plot

```python
#Pie plot to show the percentage of "yes" and "no" of card att. in dataset
#counting of yes and no in card column
yes=data[data.card =='yes'][['card']].count()
print(yes)
no=data[data.card =='no'][['card']].count()
print(no)
slices=[yes,no]
activities= ['Yes','No']
cols=['c','m']
plt.pie(slices,
       labels=activities,
       colors=cols,
       startangle=90,
       shadow=True,
       explode=(0,0.1),
       autopct='%1.1f%%')

plt.title('Pie Plot (Card)')
plt.show()
#Here, it shows that 77.6% of data is "yes" that means got approved for credit card acceptance
#and 22.4% of data is "no" that means they got rejected for credit card approval.
```

##### choosing the feature attributes

```python

X=data[["reports","expenditure","age","income","share"]]
y=data["card"]
```

```python
print(X.describe())
```

#### splitting the train and test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
#length of train set
len(X_train)
```

```python
#length of test set
len(X_test)
```

```python
# displaying train set
X_train.head()
```

```python
#displaying the test set
X_test.head()
```

### applying Random Forest Classifier

```python
#create a random forest classifier.By convention,clf maeans "classifier"
clf=RandomForestClassifier(n_jobs=2,random_state=0)
#train the classifier to take the training features and learn how they relate to the training y(the species)=fit()
clf.fit(X_train,y_train)#passing x and y(realting attributes to diff species)
#n_jobs=2(allocate 2 cpu's if available to perform job):::
```

```python
#apply the classifier we trained to the test data(which,remember,it has never seen b4)
clf.predict(X_test) # test matching
#pssing x and getting y in predict func
```

```python
#predict the probability of 1st 10 obs
clf.predict_proba(X_test)[0:10]
```

```python
# displaying predicted data for test dataset
preds= clf.predict(X_test)
preds[0:10]
```

```python
# displaying actual result of test dataset
y_test.head(10)
```

### creating a confusion matrix

```python

pd.crosstab(y_test,preds,rownames=["Actual result"],colnames=["Predicted result"])
```

### score of each attribute

```python
#view a list of the features and theirv imoportant score
list(zip(X_train,clf.feature_importances_))
#which attributes are more  decisive...used in overfitting to drop columns
```

### Efficiency of the model

```python
eff=clf.score(X_test,y_test)
```
